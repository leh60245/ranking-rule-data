```python
import os

def get_files_without_extensions(folder, ignore_extensions=None):
    if ignore_extensions is None:
        ignore_extensions = ['.json', '.jpg']

    def remove_extension(file_name):
        for ext in ignore_extensions:
            if file_name.lower().endswith(ext.lower()):
                return file_name[: -len(ext)]
        return file_name

    return {remove_extension(file) for file in os.listdir(folder)}

def compare_folders(folder1, folder2, ignore_extensions=None):
    # 폴더1의 파일 목록 가져오기 (확장자 제거)
    files_in_folder1 = get_files_without_extensions(folder1, ignore_extensions)
    # 폴더2의 파일 목록 가져오기 (확장자 제거)
    files_in_folder2 = get_files_without_extensions(folder2, ignore_extensions)
    print(len(files_in_folder1))
    print(len(files_in_folder2))
    # 폴더1에만 있는 파일 찾기
    only_in_folder1 = files_in_folder1 - files_in_folder2
    # 폴더2에만 있는 파일 찾기
    only_in_folder2 = files_in_folder2 - files_in_folder1

    # 결과 출력
    if only_in_folder1:
        print(f"폴더1에만 있는 파일 ({len(only_in_folder1)}):")
        for file in only_in_folder1:
            print(file)
    else:
        print("폴더1에는 폴더2와 동일한 파일이 있습니다.")

    if only_in_folder2:
        print(f"폴더2에만 있는 파일 ({len(only_in_folder2)}):")
        for file in only_in_folder2:
            print(file)
    else:
        print("폴더2에는 폴더1과 동일한 파일이 있습니다.")

# 사용 예시
folder1_path = '/home/leh/ranking-rule-data/data/test'
folder2_path = '/data/jwsuh/construction/sampled_test'
compare_folders(folder1_path, folder2_path)
```
