```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data/train-roberta-score.xlsx", sheet_name=[2])[2]
```

```python
plot_df = df.loc[[7,9,10,11,12]]
plot_df.set_index('Learning Rate', inplace=True)
x = plot_df.columns[2:12]

plt.figure(figsize=(12, 8))
for model in plot_df.index:
    plt.plot(x, plot_df.loc[model][2:12], marker='o', label=model)

# 그래프 제목 및 레이블 설정
plt.title('Accuracy Comparison')
plt.xlabel('Top-N')
plt.ylabel('Accuracy')
plt.legend(title='Learning Rate')
plt.grid(True)

# 그래프 표시
plt.show()
```
